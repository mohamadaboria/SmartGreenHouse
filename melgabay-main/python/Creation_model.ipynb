### What this notebook does

- Loads the `plant_village` dataset using `tfds.load()`
- Visualizes a few sample images and their labels
- Preprocesses images (resizing to 128x128, normalizing)
- Splits the dataset into training and validation sets
- Defines 3 CNN architectures:
  - A simple CNN
  - A deeper CNN
  - A CNN with Dropout layers to reduce overfitting
- Compiles the models using the Adam optimizer and sparse categorical crossentropy
- Prepares for training (batches, prefetches, splits)

This structure allows you to compare the performance of different CNN architectures on the same dataset under the same training conditions.

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split

# Load the PlantVillage dataset
def load_data():
    dataset, info = tfds.load("plant_village", with_info=True, as_supervised=True)
    return dataset['train'], info

train_data, info = load_data()

# Visualisation of few images
def show_sample_images(dataset, num_samples=3):
    plt.figure(figsize=(10, 10))
    for i, (image, label) in enumerate(dataset.take(num_samples)):
        plt.subplot(1, num_samples, i+1)
        plt.imshow(image)
        plt.title(info.features['label'].int2str(label))
        plt.axis('off')
    plt.show()

show_sample_images(train_data)

# Image preprocessing (resizing, normalization)
def preprocess_image(image, label):
    image = tf.image.resize(image, [128, 128])  # Redimensionner l'image
    image = tf.cast(image, tf.float32) / 255.0  # Normaliser l'image
    return image, label

# Apply preprocessing
train_data = train_data.map(preprocess_image)
train_data = train_data.batch(32).prefetch(tf.data.experimental.AUTOTUNE)

# Split data into train and validation
train_size = int(0.8 * len(train_data))
val_size = len(train_data) - train_size

train_dataset = train_data.take(train_size)
val_dataset = train_data.skip(train_size)

# Create CNN model
def create_cnn_model():
    model = models.Sequential([
        layers.InputLayer(input_shape=(128, 128, 3)),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dense(38, activation='softmax')  # 38 classes
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Create a deeper CNN model
def create_deep_cnn_model():
    model = models.Sequential([
        layers.InputLayer(input_shape=(128, 128, 3)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(256, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(1024, activation='relu'),
        layers.Dense(38, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Create a model with Dropout to avoid overfitting
def create_cnn_with_dropout():
    model = models.Sequential([
        layers.InputLayer(input_shape=(128, 128, 3)),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.5),  # Dropout to avoid overfitting
        layers.Dense(38, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Train and evaluate a model
def train_and_evaluate_model(model, train_dataset, val_dataset, model_name):
    history = model.fit(train_dataset, validation_data=val_dataset, epochs=10)
    # Enregistrer le modle aprs l'entranement
    model.save(f"plant_village_{model_name}.h5")
    return history

# Create and train models
models = {
    "CNN": create_cnn_model(),
    "Deep CNN": create_deep_cnn_model(),
    "CNN with Dropout": create_cnn_with_dropout(),
}

results = {}

for model_name, model in models.items():
    print(f"Entranement du modle {model_name}...")
    history = train_and_evaluate_model(model, train_dataset, val_dataset, model_name)
    results[model_name] = history.history

# Plot accuracy and loss curves
def plot_results(results):
    plt.figure(figsize=(12, 6))

    for model_name, history in results.items():
        plt.subplot(1, 2, 1)
        plt.plot(history['accuracy'], label=f'{model_name} - Train')
        plt.plot(history['val_accuracy'], label=f'{model_name} - Validation')
        plt.title('Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')

        plt.subplot(1, 2, 2)
        plt.plot(history['loss'], label=f'{model_name} - Train')
        plt.plot(history['val_loss'], label=f'{model_name} - Validation')
        plt.title('Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')

    plt.legend()
    plt.show()

plot_results(results)
```
